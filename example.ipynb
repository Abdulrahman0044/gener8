```python
import pandas as pd
from gener8 import Gener8Engine, Evaluator

# Initialize engine
engine = Gener8Engine()

# Load data
data = pd.read_csv(r"C:\Users\Abdulrahman A\OneDrive\Documents\Downloads\flat-training (original).csv")
```

```python
data.head()
```

```python
# Load and train
engine.load_and_train(    
    data,
    n_components=3,
    epochs=5,
    max_sample_size=10000,
    batch_size=32,
    gradient_accumulation_steps=2,
    max_training_time=3600,
    max_epochs=100,
    max_sequence_window=1,
    enable_flexible_generation=True,
    value_protection=True,
    rare_category_replacement_method="CONSTANT",
    differential_privacy=None)
```

```python
# Generate synthetic data
synthetic_data = engine.generate(100)
```

```python
# Evaluate
evaluator = Evaluator()
results = evaluator.evaluate(data, synthetic_data, output_dir="visualizations")
print(f"Average Univariate Accuracy: {results['avg_univariate_accuracy']:.4f}")
print(f"Average Bivariate Accuracy: {results['avg_bivariate_accuracy']:.4f}")
print(f"Average Trivariate Accuracy: {results['avg_trivariate_accuracy']:.4f}")
print(f"NNDR: {results['nndr']:.4f}")
print(f"DCR: {results['dcr']:.4f}")
```
